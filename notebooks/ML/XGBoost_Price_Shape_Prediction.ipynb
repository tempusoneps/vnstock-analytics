# 🤖 XGBoost Machine Learning Report: Phân Loại `price_shape` (12 Dạng Hình Thái Phiên)

## 📌 Mục Tiêu Phân Tích & Bối Cảnh
Notebook này thực hiện huấn luyện, phân tích hiệu năng và phân loại mô hình **`price_shape`** (12 dạng cấu trúc giá trong phiên giao dịch VN30F1M: *strong_bull, strong_bear, bull_grind, bear_grind, bull_trap, bear_trap, bull_reversal, bear_reversal, narrow_range, two_sided_bull, two_sided_bear, two_sided_neutral*).

### 🎯 Các kịch bản so sánh:
1. **Random Shuffle Split (80/20)**: Tái lập kết quả báo cáo ban đầu `report.md` (Balanced Accuracy ~60.58%).
2. **Time-Series Split (All Features ~658)**: Đánh giá theo chuỗi thời gian out-of-time.
3. **Time-Series Split (Top 20 Features)**: Tối ưu hóa mô hình với 20 đặc trưng hàng đầu.


```python
# 1. Import Thư Viện
import sys, os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, balanced_accuracy_score, f1_score
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 120
print('✔ Thư viện đã sẵn sàng!')
```

```python
# 2. Load & Clean Dataset
feature_path = '/home/v002353/Desktop/SN/04. TPO/fl-data-profiler/datasets/feature.parquet'
label_path = '/home/v002353/Desktop/SN/04. TPO/fl-data-profiler/datasets/label.csv'

features = pd.read_parquet(feature_path)
labels = pd.read_csv(label_path)

if 'Date' in features.columns:
    features['Date'] = pd.to_datetime(features['Date'])
    labels['Date'] = pd.to_datetime(labels['Date'])
    df = pd.merge(features, labels, on='Date', how='inner')
else:
    features.index = pd.to_datetime(features.index)
    labels['Date'] = pd.to_datetime(labels['Date'])
    df = features.merge(labels, left_index=True, right_on='Date', how='inner')

target_col = 'price_shape'
df_clean = df.dropna(subset=[target_col]).copy()

all_label_cols = ['allow_entry', 'remain_session_volatility', 'remain_session_net_move', 
                  'long_mae', 'short_mae', 'long_mfe', 'short_mfe', 'rr_long', 'rr_short', 
                  'eod_return_long', 'eod_return_short', 'price_range', 'intraday_position', 'price_shape']
drop_metadata = ['Date', 'trade_date', 'session_start_time', 'session_end_time']
drop_cols = [c for c in all_label_cols if c in df_clean.columns] + [c for c in drop_metadata if c in df_clean.columns]

y = df_clean[target_col]
X_raw = df_clean.drop(columns=drop_cols, errors='ignore')
numeric_cols = X_raw.select_dtypes(include=[np.number]).columns.tolist()
X = X_raw[numeric_cols].replace([np.inf, -np.inf], np.nan)
nan_ratio = X.isnull().mean()
valid_features = nan_ratio[nan_ratio < 0.5].index.tolist()
X = X[valid_features].fillna(X.median())

le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = list(le.classes_)
print(f'✔ Số lượng lớp ({len(class_names)}): {class_names}')
print(f'✔ Features: {X.shape[1]}, Rows: {len(X):,}')
```

## 🔀 Kịch Bản 1: Random Shuffle Split (Tái Lập `report.md`)

```python
# 3. Random Shuffle Split (20k rows sampling như report.md)
df_20k = df_clean.sample(n=min(20000, len(df_clean)), random_state=42)
y_20k = le.transform(df_20k[target_col])
X_20k = df_20k[numeric_cols].replace([np.inf, -np.inf], np.nan)[valid_features].fillna(X.median())

x_tr_r, x_te_r, y_tr_r, y_te_r = train_test_split(X_20k, y_20k, test_size=0.2, random_state=42, stratify=y_20k)

model_rand = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.08, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
model_rand.fit(x_tr_r, y_tr_r)

p_tr_r = model_rand.predict(x_tr_r)
p_te_r = model_rand.predict(x_te_r)

acc_r = accuracy_score(y_te_r, p_te_r)
b_acc_r = balanced_accuracy_score(y_te_r, p_te_r)
f1_w_r = f1_score(y_te_r, p_te_r, average='weighted')

print('='*50)
print('📌 Random Shuffle Split (report.md benchmark):')
print(f'   Accuracy         : {acc_r:.4f}')
print(f'   Balanced Accuracy: {b_acc_r:.4f}  <-- Khớp report.md (~0.6058)')
print(f'   Weighted F1      : {f1_w_r:.4f}')
print('='*50)
```

## ⏱️ Kịch Bản 2: Time-Series Split (All Features vs Top 20)

```python
# 4. Time-Series Chronological Split
split_idx = int(len(X) * 0.8)
X_train_ts, X_test_ts = X.iloc[:split_idx], X.iloc[split_idx:]
y_train_ts, y_test_ts = y_encoded[:split_idx], y_encoded[split_idx:]

model_ts_all = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.08, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
model_ts_all.fit(X_train_ts, y_train_ts)
p_te_ts_all = model_ts_all.predict(X_test_ts)

acc_ts_all = accuracy_score(y_test_ts, p_te_ts_all)
b_acc_ts_all = balanced_accuracy_score(y_test_ts, p_te_ts_all)
f1_w_ts_all = f1_score(y_test_ts, p_te_ts_all, average='weighted')

# Top 20 Features
imp = pd.Series(model_ts_all.feature_importances_, index=X.columns).sort_values(ascending=False)
top20_cols = imp.head(20).index.tolist()

model_ts_20 = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.08, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
model_ts_20.fit(X_train_ts[top20_cols], y_train_ts)
p_te_ts_20 = model_ts_20.predict(X_test_ts[top20_cols])

acc_ts_20 = accuracy_score(y_test_ts, p_te_ts_20)
b_acc_ts_20 = balanced_accuracy_score(y_test_ts, p_te_ts_20)
f1_w_ts_20 = f1_score(y_test_ts, p_te_ts_20, average='weighted')

print('='*50)
print('📌 Time-Series Split (All Features):')
print(f'   Accuracy: {acc_ts_all:.4f} | Balanced Acc: {b_acc_ts_all:.4f} | Weighted F1: {f1_w_ts_all:.4f}')
print('📌 Time-Series Split (Top 20 Features):')
print(f'   Accuracy: {acc_ts_20:.4f} | Balanced Acc: {b_acc_ts_20:.4f} | Weighted F1: {f1_w_ts_20:.4f}')
print('='*50)
```

```python
# 5. Confusion Matrix Visualizations
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
cm_all = confusion_matrix(y_test_ts, p_te_ts_all, normalize='true')
cm_20 = confusion_matrix(y_test_ts, p_te_ts_20, normalize='true')

sns.heatmap(cm_all, annot=True, fmt='.2f', cmap='Blues', ax=axes[0], xticklabels=class_names, yticklabels=class_names, cbar=False)
axes[0].set_title(f'Confusion Matrix: All Features (Balanced Acc = {b_acc_ts_all:.2f})', fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)

sns.heatmap(cm_20, annot=True, fmt='.2f', cmap='Greens', ax=axes[1], xticklabels=class_names, yticklabels=class_names, cbar=False)
axes[1].set_title(f'Confusion Matrix: Top 20 Features (Balanced Acc = {b_acc_ts_20:.2f})', fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 6. Bảng Benchmark Tổng Hợp
bench_df = pd.DataFrame({
    'Kịch Bản': ['1. Random Shuffle (report.md)', '2. Time-Series (All Feats)', '3. Time-Series (Top 20 Feats)'],
    'Split': ['Random Shuffle (Data Leakage)', 'Chronological Out-of-Time', 'Chronological Out-of-Time'],
    'Accuracy': [round(acc_r, 4), round(acc_ts_all, 4), round(acc_ts_20, 4)],
    'Balanced Accuracy': [round(b_acc_r, 4), round(b_acc_ts_all, 4), round(b_acc_ts_20, 4)],
    'Weighted F1': [round(f1_w_r, 4), round(f1_w_ts_all, 4), round(f1_w_ts_20, 4)]
})
display(bench_df)
```
